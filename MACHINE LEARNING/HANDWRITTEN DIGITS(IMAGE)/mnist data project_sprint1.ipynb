# a. Identify the Target Variable and Splitting the Data into Train and Test

```python
y=df0['index'] #target variable

X=df0.iloc[:,0:784] #input variable
```

```python
#our target variable is discrete(nominal),so we are using CLASIIFICATION TASK
#for CLASIIFICATION task:ACCURACY
```

```python
# split into train and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=100)

X_train.head()
```

# STEP-4 DATA PREPARATION ON TRAIN DATA

# Separating Categorical and Numerical Columns:


```python
X_train.head()
```

```python
X_train.dtypes
```

```python
#NUMERICAL COLUMNS
X_train_num = X_train.select_dtypes(include=['float64'])

X_train_num.head()
```

```python
X_train_num.shape
```

# Scaling(standarization) the Numerical Features in the train data

```python
X_train_num.head()
```

```python
# scaling the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

# STEP-5 DATA PREPARATION ON TEST DATA

# Preparing Test Data


```python
X_test.head()
```

```python
X_test.dtypes
```

```python
#NUMERICAL COLUMNS
X_test_num = X_test.select_dtypes(include=['float64'])

X_test_num.head()
```

# Scaling the Numerical Features

```python
X_test_num_rescaled = pd.DataFrame(scaler.transform(X_test_num), 
                                    columns = X_test_num.columns, 
                                    index = X_test_num.index)

X_test_num_rescaled.head()
```

# STEP-6 use algorithms here:

# LOGISTIC REGRESSION

```python
#for test prediction:
```

```python
from sklearn.linear_model import LogisticRegression
regressor= LogisticRegression()
regressor.fit(X_train_num_rescaled, y_train)
```

```python
y_test_pred = regressor.predict(X_test_num_rescaled)
```

```python
temp_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

temp_df.head()
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test, y_test_pred)
```
